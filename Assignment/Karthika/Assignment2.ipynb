Outlier detection and removal

```python
sns.boxplot(x=df['Age'])
```

```python
print('skewness value of Age: ',df['Age'].skew())
print('skewness value of Fare: ',df['CustomerId'].skew())
```

```python
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1
whisker_width = 1.5
Age_outliers = df[(df['Age'] < Q1 - whisker_width*IQR) | (df['Age'] > Q3 + whisker_width*IQR)]
Age_outliers.head()
```

```python
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1
whisker_width = 1.5
l = Q1 -(whisker_width*IQR)
u = Q3 + (whisker_width*IQR)
df['Age']=np.where(df['Age']>u,u,np.where(df['Age']<l,l,df['Age']))
```

```python
sns.boxplot(x=df['Age']) #outliesrs removed
```
